```python
import requests
from dotenv import load_dotenv
import os
from geopy.geocoders import Nominatim

load_dotenv()
API_KEY = os.getenv("OPENWEATHER_API_KEY")
print(API_KEY)
```

```python
headers = {
    "User-Agent": "MyWeatherApp/1.0 puskarpy"
}
```

```python
# Get latitude and Longitude
def GetLatAndLon(place):
    geolocator = Nominatim(user_agent="weather-project")
    location = geolocator.geocode(place)
    if location:
        print(location.address)
        return location.latitude, location.longitude
    else:
        return None, None

# Get all weather info 
def GetWeather(place):
    lat, lon = GetLatAndLon(place)
    if lat == None:
        return None
        
    url = f"https://api.met.no/weatherapi/locationforecast/2.0/compact?lat={lat}&lon={lon}"
    res = requests.get(url, headers=headers)
    if res.status_code == 200:
        return res.json()
    else:
        print(f"Error {res.ststus_code}")
        print(res.text)
        return None

# Get infos with units
def GetInfoAndUnit(data, param):
    try:
        details = data['properties']['timeseries'][0]['data']['instant']['details']
        units = data['properties']['meta']['units']
        return f"{details[param]} {units[param]}"
    except KeyError:
        return f"{param} not found in data"
```

```python
data = GetWeather("Kathmandu")
```

```python
if data:
    temp = GetInfoAndUnit(data, "air_temperature")
    humidity = GetInfoAndUnit(data, "relative_humidity")
    print("Temperature:", temp)
    print("Humidity:", humidity)
```
